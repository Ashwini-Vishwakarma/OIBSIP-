```python
df["Age"].describe()

df["Age"] = df["Age"].fillna(df["Age"].median())
```

### Missing Value Handling – Age

The `Age` column contained missing values.

Since it is a numerical feature and may contain outliers, the missing values were replaced using the **median** instead of the mean. The median is more robust to extreme values.

```python
df = df.drop(columns=["Cabin"])
```

### Missing Value Handling – Cabin

The `Cabin` column contained more than 75% missing values.

Since such a large portion of the data was unavailable, the column was removed because imputing values would not produce reliable information.

```python
print(df["Embarked"].mode())

df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

### Missing Value Handling – Embarked

The `Embarked` column had only two missing values.

Since it is a categorical feature, the missing values were replaced using the **mode** (most frequent value).

```python
print(df.isnull().sum())
```

## Observation

- Missing values in the `Age` column were filled using the median.
- The `Cabin` column was removed due to excessive missing values.
- Missing values in the `Embarked` column were filled using the mode.
- After cleaning, the dataset contains no missing values.

```python
duplicates_before = df.duplicated().sum()

print("Duplicate rows before cleaning:", duplicates_before)

df = df.drop_duplicates()

duplicates_after = df.duplicated().sum()

print("Duplicate rows after cleaning:", duplicates_after)

removed_duplicates = duplicates_before - duplicates_after

print("Number of duplicate rows removed:", removed_duplicates)
```

## Standardization

### Date Standardization

The Titanic dataset does not contain any date-related columns. Therefore, no date conversion using `pd.to_datetime()` was required.

```python
Q1 = df["Fare"].quantile(0.25)
Q3 = df["Fare"].quantile(0.75)

print("Q1:", Q1)
print("Q3:", Q3)

IQR = Q3 - Q1

print("IQR:", IQR)

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

print("Lower Limit:", lower_limit)
print("Upper Limit:", upper_limit)

outliers = df[
    (df["Fare"] < lower_limit) |
    (df["Fare"] > upper_limit)
]

print(outliers)

print("Number of Outliers:", outliers.shape[0])
```

## Outlier Detection – Fare

The IQR method was used to detect outliers in the `Fare` column.

Although several outliers were identified, they were **retained** because they represent genuine differences in ticket prices rather than incorrect data entries. Removing them could result in the loss of valuable information.
